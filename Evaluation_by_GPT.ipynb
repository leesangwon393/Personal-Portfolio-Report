```python
import pandas as pd
import openai
api_key = "YOUR_API_KEY_HERE"
client = openai.OpenAI(api_key=api_key)
```

```python
outputs = [" "] # 우리의 결과
prompt = """You are a financial report evaluator.
You will receive one portfolio report text.
Evaluate it strictly according to the following criteria.

For each item:
- Assign an integer score from 1 to 5 (1 = very poor, 3 = average, 5 = excellent).
- Add one concise reason (≤15 words) explaining the score.
- Avoid vague words like "somewhat" or "a bit".
- Be consistent across all items.

Evaluation Criteria:
1. Accuracy (Are financial/news data factually correct?)
2. Consistency & Logic (Are data and conclusions logically connected?)
3. Suitability to Investor Profile (Does it fit Aggressive / Neutral / Conservative investor type?)
4. Risk Awareness (Does it address investment risks in a balanced way?)
5. Numerical Interpretation (Does it interpret financial metrics correctly?)
6. Explainability (Does it explain the reasoning behind conclusions?)
7. Professionalism & Readability (Is the tone professional and easy to read?)

Final Score = average of all numeric scores (rounded to one decimal place).

Output Format (use exactly this format):
Accuracy: 4 (EPS data matches, but last quarter revenue is missing)
Consistency & Logic: 5 (Strong logical link between data and conclusion)
Suitability to Investor Profile: 3 (Investor type is mentioned but not fully developed)
Risk Awareness: ...
...
Final Score: 4.1 / 5

"""
```

```python
def generate_response(full_prompt):
    messages = [
        {"role": "system", "content": "You are a strict financial report evaluator who scores objectively."},
        {"role": "user", "content": full_prompt}
    ]
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=messages,
        temperature=0,
        max_tokens=300
    )
    return response.choices[0].message.content.strip()
```

```python
result = []
for n in outputs:
    p = prompt + "\n\nPortfolio Report:\n" + n
    response = generate_response(p)
    result.append(response)

print(result[0])
```
